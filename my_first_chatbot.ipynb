```python
# chatbot
def chatbot():
    print("Bot:hello! type 'bye' to exit.")

    while True:
        user = input("You: ").lower()



        if "hi" in user:
            print("Bot:hello")

        elif "how are you" in user:
            print("Bot: I'm fine,thanks!")

        elif "your name" in user:
            print("Bot: I am Your chatbot")

        elif "help" in user:
            print("Bot: I can respond to greeting and basic questions.")

        elif "bye" in user:
            print("Bot: Goodbye")
            break
        else:
            print("Bot:Sorry, I don't understand")
    
chatbot()    
```
